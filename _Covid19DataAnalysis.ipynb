Import Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print('Modules are imported.')
```

Import Covid-19 Dataset


```python
corona_dataset_csv=pd.read_csv('covid19_Confirmed_dataset.csv')
corona_dataset_csv.head(10)
```

Lets check shape of the dataframe


```python
corona_dataset_csv.shape
```

Delete the useless columns in our data frame


```python
corona_dataset_csv.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
corona_dataset_csv.head(10)
```

Aggregating the Rows by Country

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.shape
```

Visualising Data related to a Country

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

Calculating a good measure

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

calculating and plotting the first derivative of the curve

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

Find maximum infection rate for specific Countries

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

Maximum infection rate for all Countries

```python
countries=list(corona_dataset_aggregated.index)
max_infection_rate=[]
for c in countries :
  max_infection_rate.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"]=max_infection_rate
```

```python
corona_dataset_aggregated.head()
```

Creating new data frame with only columns needed

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

Importing The Dataset

```python
happiness_report_csv=pd.read_csv('worldwide_happiness_report.csv')
```

```python
happiness_report_csv.head(10)
```

Drop Useless columns

```python
useless_cols=["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report_csv.drop(useless_cols, axis=1, inplace=True)
```

```python
happiness_report_csv.head()
```

Changing the indices of the DataFrame

```python
happiness_report_csv.set_index("Country or region", inplace=True)
```

```python
happiness_report_csv.head()
```

**Now lets join the Two Data Sets**

Corona Dataset

```python
corona_data.head()
```

```python
corona_data.shape
```

Happiness Report dataset

```python
happiness_report_csv.head()
```

```python
happiness_report_csv.shape
```

```python
data = corona_data.join(happiness_report_csv,how="inner")
data.head()
```

**Correlation Matrix**

```python
data.corr()
```

**Visualisation of the results**

```python
data.head()
```

PLOT GDP VS MAX INFECTION RATE

```python
x=data["GDP per capita"]
y=data["max_infection_rate"]
sns.scatterplot(x,y)
```

```python
sns.regplot(x,np.log(y))
```

PLOT SOCIAL SUPPORT VS MAX INFECTION RATE

```python
x=data["Social support"]
y=data["max_infection_rate"]
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))
```
